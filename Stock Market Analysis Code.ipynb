#### Using RNN

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
x = df[['Date']]
```

```python
x.head()
```

```python
y = df[['Close']].values
```

```python
x.set_index('Date', inplace=True)
```

```python
#data = y.values.reshape(-1, 1)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(y)
```

```python
# creating sequences of seq_length size
def create_sequences(data, seq_length):
    sequences = []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i + seq_length])
    return np.array(sequences)
```

```python
seq_length = 50
sequences = create_sequences(scaled_data, seq_length)
X = sequences[:, :-1]  # All but the last element in each sequence
y = sequences[:, -1]   # The last element in each sequence (the target value)
```

```python
# Reshape X to be 3-dimensional as required by RNN: (num_samples, seq_length, num_features)
#X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# Split into training and testing sets
split = int(len(X) * 0.8)  # 80% for training, 20% for testing
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
model = Sequential()
```

```python
model.add(LSTM(units=50, return_sequences=True, input_shape=(seq_length, 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test,y_test), verbose=2)
```

```python
# Make predictions
y_pred2 = model.predict(X_test)
y_pred2 = scaler.inverse_transform(y_pred2) # Inverse transform to get actual price scale
y_test_inverse = scaler.inverse_transform(y_test)
```

```python
test_loss = model.evaluate(X_test, y_test)
print(f'Test Loss of LSTM model: {test_loss}')
```

```python
mse2=mean_squared_error(y_test,y_pred2)
mae2=mean_absolute_error(y_test,y_pred2)
```

```python
print(f" mean squared error={mse2} \n mean absolute error={mae2}")
```

```python
# Plot the results
plt.figure(figsize=(14, 5))
plt.plot(x.index[-len(y_test):], y_test_inverse,color='b', label='Actual Prices')
plt.plot(x.index[-len(y_pred2):], y_pred2, color = 'r',label='Predicted Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price Prediction')
plt.legend()
plt.show()
```

```python
# Prepare data for future prediction
initial_sequence = scaled_data[-seq_length:]  # Last seq_length data from scaled_data
initial_sequence = initial_sequence.reshape((1, seq_length, 1))  # Reshape for the RNN

# Number of future days to predict
num_future_days = 30

# Store predictions
future_predictions = []

# Generate predictions
current_sequence = initial_sequence
for _ in range(num_future_days):
    next_price = model.predict(current_sequence)[0, 0]  # Predict the next price
    future_predictions.append(next_price)
    
    # Reshape next_price to match the shape expected by np.append
    next_price = np.array([[next_price]]).reshape((1, 1, 1))
    
    # Update the current sequence
    current_sequence = np.append(current_sequence[:, 1:, :], next_price, axis=1)

# Inverse transform the predictions to get actual price scale
future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Create a date range for the future predictions
future_dates = pd.date_range(start=df['Date'].iloc[-1], periods=num_future_days + 1, inclusive='right')
```

```python
future_df = pd.DataFrame(data=future_predictions, index=future_dates, columns=['Close'])


# Reorder columns for mplfinance
future_df = future_df[['Close']]
future_df.index.name = 'Date'

# Ensure historical data is also in the same format
historical_df = df[['Close','Date']].copy()
historical_df['Date'] = pd.to_datetime(historical_df['Date'])

# Set 'Date' as the index
historical_df.set_index('Date', inplace=True)

# Ensure both DataFrames have DatetimeIndex
historical_df.index = pd.to_datetime(historical_df.index)
future_df.index = pd.to_datetime(future_df.index)

# Combine historical and future data
combined_df = pd.concat([historical_df, future_df])

```
